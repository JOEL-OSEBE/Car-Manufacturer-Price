```python
from sklearn.model_selection import train_test_split
```

```python
#split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=40)
```

Predictive models

```python
from sklearn.metrics import accuracy_score
```

#Logistic regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,confusion_matrix
```

```python
#Features to be used in Predictive Analysis
train_features=pd.DataFrame([x_train['Year'],
                            x_train['Engine Fuel Type'],
                            x_train['Engine HP'],
                            x_train['Engine Cylinders'],
                            x_train['Transmission Type'],
                            x_train['Driven_Wheels'], 
                            x_train['Number of Doors'],
                            x_train['Market Category'],
                            x_train['Vehicle Size'],
                            x_train['Vehicle Style'],
                            x_train['highway MPG'],
                            x_train['city mpg'],
                            x_train['Popularity']]).T
test_features=pd.DataFrame([x_test['Year'],
                            x_test['Engine Fuel Type'],
                            x_test['Engine HP'],
                            x_test['Engine Cylinders'],
                            x_test['Transmission Type'],
                            x_test['Driven_Wheels'], 
                            x_test['Number of Doors'],
                            x_test['Market Category'],
                            x_test['Vehicle Size'],
                            x_test['Vehicle Style'],
                            x_test['highway MPG'],
                            x_test['city mpg'],
                            x_test['Popularity']]).T
```

```python
#Logistic regression
from sklearn.metrics import classification_report
logistic=linear_model.LogisticRegression(solver="liblinear",multi_class="auto")
logistic.fit(train_features,y_train.values.ravel())
y_pred_train=logistic.predict(train_features)
y_pred_test=logistic.predict(test_features)
print("Logistic Regression:")
print("Train Accuracy Score:",accuracy_score(y_train,y_pred_train))
print("Test Accuracy Score:",accuracy_score(y_test,y_pred_test))
print("Mean Squared Error:",mean_squared_error(y_test,y_pred_test))

print("Classification Summary:")
print(classification_report(y_test, y_pred_test))
print("Confusion Matrix:")
confusion_matrix(y_test,y_pred_test)
```

```python
plt.figure(figsize=(8.5,6))
plt.title("Heatmap Predicted against Test")
plt.ylabel("Predicted")
plt.xlabel("Test")
sns.heatmap(confusion_matrix(y_test,y_pred_test),annot=True,linecolor="white",linewidth=.02,vmin=0,vmax=2500.)
```

RANDOM FOREST

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor as RF

regressor = RF(n_estimators=10,
             criterion='mse',
             max_features= None, 
             max_depth = 14)
regressor=regressor.fit(train_features, y_train.values.ravel())
regressor.fit(train_features, y_train.values.ravel())
y_pred = regressor.predict(test_features)
```

```python
from sklearn.metrics import classification_report
y_pred=pd.DataFrame(y_pred)
y_pred=y_pred.astype("int")
print("Random forest:")
print("Train Accuracy Score:",accuracy_score(y_test,y_pred))
print("Mean Squared Error:",mean_squared_error(y_test,y_pred))
print("Confusion Matrix:")
confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize=(8.5,6))
plt.title("Heatmap Predicted against Test")
plt.ylabel("Predicted")
plt.xlabel("Test")
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,linecolor="white",linewidth=.02,vmin=0,vmax=2500)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dt=tree.DecisionTreeClassifier()
dtr=dt.fit(train_features, y_train.values.ravel())
dtr.fit(train_features, y_train.values.ravel())
y_predt = dtr.predict(test_features)
```

```python
from sklearn.metrics import classification_report
y_predt=pd.DataFrame(y_predt)
y_predt=y_predt.astype("int")
y_test=pd.DataFrame(y_test)
y_test=y_test.astype("int")
print("Decision tree:")
print("Train Accuracy Score:",accuracy_score(y_test,y_predt))
print("Classification Summary:")
print(classification_report(y_test, y_predt))
print("Mean Squared Error:",mean_squared_error(y_test,y_predt))
print("Confusion Matrix:")
confusion_matrix(y_test,y_predt)
```

```python
plt.figure(figsize=(8.5,6))
plt.title("Heatmap Predicted against Test")
plt.ylabel("Predicted")
plt.xlabel("Test")
sns.heatmap(confusion_matrix(y_test,y_predt),annot=True,linecolor="white",linewidth=.02,vmin=0,vmax=2500,center=500)
```

Use Decision Tree classifier which has a higher accuracy level

```python
from sklearn.tree import export_graphviz
```

```python
import pydotplus
import collections
```

```python
# Visualize model Decision Tree

dot_data = tree.export_graphviz(dtr,
                                feature_names=test_features.columns,
                                out_file=None,
                                filled=True,
                                rounded=True,
                               special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)

colors = ('turquoise', 'orange')
edges = collections.defaultdict(list)

for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))

for edge in edges:
    edges[edge].sort()    
    for i in range(2):
        dest = graph.get_node(str(edges[edge][i]))[0]
        dest.set_fillcolor(colors[i])

graph.write_png('tree.png')
```
